```python


df1 = df[df['finish_time'] < 1800]

event_participants = df1.groupby('record_id')['athlete_id'].count()

# Фильтруем пробежки с количеством участников меньше 47
small_events = event_participants[event_participants < 47].index

# Оставляем только данные для отобранных пробежек
filtered_df = df[df['record_id'].isin(small_events)]

# Сортируем данные для анализа
filtered_df = filtered_df.sort_values(by=['record_id', 'finish_time'])

# Функция для подсчета участников с разницей <= 5 секунд в пробежке
def calculate_close_finishers(group):
    group['time_diff'] = group['finish_time'].diff().fillna(0).abs()
    group['group'] = (group['time_diff'] > 5).cumsum()
    group_counts = group.groupby('group')['athlete_id'].count()
    valid_groups = group_counts[group_counts > 1]
    return valid_groups.sum()

# Применяем функцию к каждой подходящей пробежке
result = filtered_df.groupby('record_id').apply(calculate_close_finishers)

# Общий результат по всем подходящ
result.mean()
```

```python
import plotly.graph_objects as go

# Данные
stages = ['Все участники', 'Новички', 'Новички, пробежавшие более одного раза']
values = [
    df['athlete_id'].nunique(),  # Все участники
    df[df['is_newbie'] == 1]['athlete_id'].nunique(),  # Новички
    df[(df['is_newbie'] == 1) & (df['run_count'] > 1)]['athlete_id'].nunique()  # Новички, пробежавшие более одного раза
]
formatted_values = [f"{v:,}".replace(",", " ") for v in values]  # Разделяем тысячи пробелами
percentages = ["100%"]  # Первый этап не имеет процента от предыдущего
for i in range(1, len(values)):
    percentages.append(f"{(values[i] / values[i-1] * 100):.1f}%")  # Рассчитываем проценты

text_labels = [
    f"{formatted_values[i]}<br>{percentages[i] if percentages[i] else ''}" 
    for i in range(len(values))
]
# Создаем график воронки
fig = go.Figure(go.Funnel(
    y=stages,  # Этапы
    x=values,  # Количество на каждом этапе
    textinfo="none",  # Подписи на каждом этапе
    texttemplate="%{text}",  # Подключаем текст из text
    text=text_labels,
    textposition="inside",  # Позиция текста внутри блоков
    marker=dict(
        color=[main_color, green, red]  # Цвета сегментов
    ),
    connector=dict(
        visible=False  # Отключаем линии между сегментами
    ),
    textfont=dict(size=28),  # Увеличиваем размер шрифта
))

# Центрируем график и убираем фон, а также убираем промежутки
fig.update_layout(
    plot_bgcolor=grey,  # Цвет фона графика
    paper_bgcolor=grey,  # Цвет фона страницы
    funnelmode='stack',  # Используем stack для плотного расположения блоков
    showlegend=False,  # Убираем легенду
    xaxis=dict(visible=False),  # Убираем ось X
    yaxis=dict(
        visible=False  # Отключаем ось Y
    ),
    margin=dict(l=100, r=100, t=50, b=50),  # Подстраиваем отступы
)

# Сохраняем график
fig.write_html("funnel_chart.html")
```
